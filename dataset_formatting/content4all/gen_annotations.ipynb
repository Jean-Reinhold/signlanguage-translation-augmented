```python
import json
import pandas as pd


DATA_PATH = "/mnt/disk3Tb/slt-datasets/Content4All/vrt-news-annotations.json"
json_data = json.load(open(DATA_PATH))
df = pd.DataFrame(json_data)
df.head()
```

```python
df_processed = df.T
df_processed["sequence_id"] = df_processed["sequence_id"].astype(str)
df_processed["video_id"] = df_processed["video_id"].astype(str)
df_processed["id"] = df_processed["video_id"] + "_" + df_processed["sequence_id"]
df_processed.drop(["video_id", "sequence_id"], axis=1, inplace=True)
df_processed.rename(columns={"annotation": "text"}, inplace=True)
df_processed = df_processed[['id', 'text', 'start_ms', 'stop_ms', 'start_frame', 'stop_frame', 'video_path', 'openpose2D_path', 'openpose-2Dto3D_path']]
df_processed.head()
```

```python
df_processed.to_csv("/mnt/disk3Tb/slt-datasets/Content4All-mirror/vrt-news-annotations.csv", index=False)
```

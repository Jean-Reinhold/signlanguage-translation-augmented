```python
import pandas as pd


DIR = "/mnt/disk3Tb/slt-datasets/LSAT"
annotations = pd.read_csv(f"{DIR}/annotations.csv")
annotations.head()
```

```python
# get distribution of playlists
annotations["playlist"].value_counts(normalize=True)
```

```python
# use train test split from sklearn to do a stratified splitting based on playlist
from sklearn.model_selection import train_test_split

train, test = train_test_split(
    annotations, test_size=0.2, stratify=annotations["playlist"]
)
# gen validation set
train, val = train_test_split(train, test_size=0.2, stratify=train["playlist"])
```

```python
train["playlist"].value_counts(normalize=True)
```

```python
test["playlist"].value_counts(normalize=True)
```

```python
val["playlist"].value_counts(normalize=True)
```

```python
annotations["split"] = "train"
annotations.loc[annotations.index.isin(test.index), "split"] = "test"
annotations.loc[annotations.index.isin(val.index), "split"] = "val"
```

```python
annotations.value_counts("split")
```

```python
annotations.to_csv(f"{DIR}/annotations.csv", index=False)
```
